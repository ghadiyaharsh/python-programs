```python
'''What is Seaborn?

Built on top of Matplotlib
Used for statistical data visualization
Works well with Pandas DataFrames
'''
```

Seaborn uses DataFrames
Default aggregation = mean
hue = grouping
Built-in datasets = tips, iris, titanic
Works with Matplotlib (plt.show())

```python
import seaborn as sns # Import Seaborn for data visualization
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
```

```python
# Load the built-in 'tips' dataset
df = sns.load_dataset('tips')
print(df.head())  # Display the first few rows of the dataset
```

```python
print(df.info())  # Display information about the dataset
```

```python
print(df.shape)
```

```python
print(df.size) # Display the total number of elements in the dataset
```

```python
# Themes / Styles
sns.set_style('whitegrid')  # Set the style to 'whitegrid'
sns.set_style('darkgrid')  # Set the style to 'darkgrid', Changes background style of plots
```

Basic Plots

```python
#1. Scatter Plot - Shows relationship between two numerical variables

sp = sns.scatterplot(x='total_bill', y='tip', data=df)
print(sp)
```

```python
#2. Line Plot -Shows trends over time or ordered data
sns.lineplot(x='size', y="total_bill", data=df)
# size of the group vs total bill
```

```python
#3. bar plot - Shows relationship between a categorical variable and a numerical variable
sns.barplot(x="day", y="total_bill", data=df)


'''
Concept:
Compares categories

Important:
default = mean value

Explanation:
Shows average bill per day
'''

```

```python
#4. Count plot - Counts frequency of categories
sns.countplot(x="day", data=df)
```

Distribution Plots

```python
#1. Histogram - show distribution of a single numerical variable
sns.histplot(df["total_bill"]) # Distribution of total bill amounts
```

```python
#2. KDE Plot - Smooth curve of distribution
sns.kdeplot(df["total_bill"])
```

```python
#3. Combined - Histogram + smooth curve together
sns.histplot(df["total_bill"], kde=True) # Histogram with KDE overlay
```

Categorical Plots

```python
#1. Box Plot - show median, quartiles, outliers
sns.boxplot(x="day", y="total_bill", data=df)
```

```python
#2. violin plot - similar to box plot but also shows density of data
sns.violinplot(x="day", y="total_bill", data=df)
```

```python
#3. Strip Plot - shows all data points
sns.stripplot(x="day", y="total_bill", data=df)
```

```python
#4. Swarm Plot - similar to strip plot but adjusts points to avoid overlap
sns.swarmplot(x="day", y="total_bill", data=df)
```

Multivariable Plots

```python
#1. Hue - it is a list of meaning of data points hue = grouping (color-based separation)
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df)
```

```python
#2. Pair Plot - show relationships between all columns
sns.pairplot(df)
```

```python
#3. Heatmap 
corr = df.corr(numeric_only = True) # Calculate correlation matrix for numeric columns
sns.heatmap(corr, annot=True) #annot=True adds correlation values to the heatmap
```

Advanced Plots

```python
#1. Facet Grid - Multiple plots based on category
g = sns.FacetGrid(df, col="time")
g.map(sns.histplot,"total_bill")
```

```python
#2. Joint Plot - Combines scatter + histogram
sns.jointplot(x="total_bill", y="tip", data=df)
```

Customization
